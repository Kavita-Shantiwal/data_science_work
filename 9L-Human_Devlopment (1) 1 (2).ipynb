## Data Visualization

### Heatmap

```python
f, ax = plt.subplots(figsize=(18, 12))
core = df.corr()
sns.heatmap(core)
```

```python
plt.figure( figsize = (15,10))
sns.lineplot(x="Country", y="HDI Rank (2021)", data=df)
```

```python
plt.figure( figsize = (15,10))
sns.lineplot(x="Country", y="GDI Group (2021)", data=df)
```
